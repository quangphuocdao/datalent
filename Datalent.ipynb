```python
#Đổi định dạng dữ liệu của cột D/O/B
from datetime import datetime, timedelta
```

```python
def final_date(value):
    if pd.notna(value):
        return (datetime(1899, 12, 30) + timedelta(days=value))
    else:
        return None

df['D/O/B'] = df['D/O/B'].apply(final_date)

# Bỏ time và giữ chỉ giữ lại ngày, xử lý NaN
df['D/O/B'] = pd.to_datetime(df['D/O/B'], errors='coerce').dt.date
```

```python
#Đổi định dạng dữ liệu của cột Saledate
df['saledate'] = pd.to_datetime(df['saledate'], errors='coerce').dt.date
```

## Exploratory Data Analysis

### 1. Thống kê tổng doanh thu và số lượng vé theo các ngày trong tháng 5/2019

```python
daily_stats = df.groupby('saledate').agg({'ticket price': ['sum', 'count']}).reset_index()

daily_stats.columns = ['saledate', 'total_ticket_price', 'ticket_count']

print(daily_stats)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(daily_stats['saledate'], daily_stats['total_ticket_price'], marker='o', label='Tổng doanh thu')
plt.title('Tổng doanh thu theo ngày')
plt.xlabel('Ngày')
plt.ylabel('Tổng doanh thu')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(daily_stats['saledate'], daily_stats['ticket_count'], marker='o', label='Số lượng vé')
plt.title('Số lượng vé bán được theo ngày')
plt.xlabel('Ngày')
plt.ylabel('Số lượng vé')
plt.legend()
plt.show()
```

### 2. Thống kê tổng doanh thu và số lượng vé được mua dựa theo thời gian bộ phim được chiếu:

```python
time_stats = df.groupby('time').agg({'ticket price': ['sum', 'count']}).reset_index()

time_stats.columns = ['time', 'total_ticket_price', 'ticket_count']

print(time_stats)
```

```python
fig = px.treemap(time_stats, 
                 path=['time'], 
                 values='total_ticket_price',
                 color='ticket_count',
                 color_continuous_scale='Plasma',
                 title='Thống kê doanh thu và số lượng vé được bán ra dựa theo thời gian chiếu của phim')

fig.show()
```

### 3. Thống kê tổng doanh thu và số lượng vé được mua dựa theo nhóm tuổi của khách hàng

```python
#Thống kê tổng doanh thu giựa trên nhóm tuổi:
bins = [0, 10, 20, 30, 40, 50, 55]
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-55']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
df['Age_Group'] = df['Age_Group'].cat.add_categories('Unknown').fillna('Unknown')

age_grouped = df.groupby('Age_Group').agg({'ticket price': ['sum', 'count']}).reset_index()
age_grouped.columns = ['Age_Group', 'Tổng doanh thu', 'Số lượng vé']

print(age_grouped)
```

```python
#Trực quan hóa bằng biểu đồ
plt.figure(figsize=(12, 6))

ax = sns.barplot(x='Age_Group', y='Tổng doanh thu', data=age_grouped)

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    pos_x = p.get_x() + width / 2 
    pos_y = height + 10  
    plt.text(pos_x, pos_y, f'{height:.0f}', ha='center', va='bottom', color='black', fontsize=12)

plt.title('Tổng doanh thu theo nhóm tuổi của khách hàng')
plt.xlabel('Nhóm tuổi')
plt.ylabel('Tổng doanh thu')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

```python
plt.figure(figsize=(12, 6))

ax = sns.barplot(x='Age_Group', y='Số lượng vé', data=age_grouped)

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    pos_x = p.get_x() + width / 2 
    pos_y = height + 10  
    plt.text(pos_x, pos_y, f'{height:.0f}', ha='center', va='bottom', color='black', fontsize=12)

plt.title('Số lượng vé được mua theo nhóm tuổi của khách hàng')
plt.xlabel('Nhóm tuổi')
plt.ylabel('Tổng doanh thu')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

### 8. Thống kê doanh thu và số lượng vé được bán ra của từng thể loại phim:

### 13. Thống kê tổng doanh thu dựa theo giới tính và thể loại phim

```python
bins = [0, 10, 20, 30, 40, 50, 55]
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-55']

df[['Genre1', 'Genre2', 'Genre3']] = df['film_type'].str.split(',', expand=True)
df['ticket price'] = pd.to_numeric(df['ticket price'], errors='coerce')
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

new_df = df[['Age_Group', 'Genre1', 'Genre2', 'Genre3', 'ticket price']]
new_df['Age_Group'] = new_df['Age_Group'].cat.add_categories('Unknown').fillna('Unknown')
new_df = pd.melt(new_df, id_vars=['Age_Group', 'ticket price'], value_vars=['Genre1', 'Genre2', 'Genre3'], var_name='Genre', value_name='Genre_Type')
new_df = new_df.dropna(subset=['Genre_Type'])

gender_age_total_price = new_df.groupby(['Age_Group', 'Genre_Type'])['ticket price'].sum().reset_index()
print(gender_age_total_price)
```

```python
# Vẽ biểu đồ
plt.figure(figsize=(12, 8))
sns.barplot(y='Age_Group', x='ticket price', hue='Genre_Type', data=gender_age_total_price, ci=None, orient='h')
plt.title('Tổng doanh thu theo nhóm tuổi và loại phim (Tổng ticket price)')
plt.xlabel('Tổng doanh thu (ticket price)')
plt.ylabel('Nhóm Tuổi')
plt.legend(title='Loại Phim')
plt.show()
```

### 14. Thống kê tổng doanh thu dựa vào giới tính và thể loại phim 

```python
df['film_type'] = df['film_type'].astype(str)
df[['Genre1', 'Genre2', 'Genre3']] = df['film_type'].str.split(',', expand=True)

df['ticket price'] = pd.to_numeric(df['ticket price'], errors='coerce')

df_selected = df[['Gender', 'Genre1', 'Genre2', 'Genre3', 'ticket price']]

gender_film_total_price = pd.melt(df_selected, id_vars=['Gender', 'ticket price'], value_vars=['Genre1', 'Genre2', 'Genre3'], var_name='Genre', value_name='Genre_Type')

gender_film_total_price = gender_film_total_price.groupby(['Gender', 'Genre_Type'])['ticket price'].sum().reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='Gender', y='ticket price', hue='Genre_Type', data=gender_film_total_price, ci=None)
plt.title('So sánh doanh thu theo giới tính và thể loại phim (Tổng ticket price)')
plt.xlabel('Giới tính')
plt.ylabel('Tổng doanh thu (ticket price)')
plt.legend(title='Thể loại phim')
plt.show()
```
